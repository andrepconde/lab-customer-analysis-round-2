```python
data ["state"] = data["state"].fillna('Unknown')
data ["response"] = data ["response"].fillna("Unknown")
data ["months_since_last_claim"] = data ["months_since_last_claim"].fillna("Unknown") 
data ["number_of_open_complaints"] = data ["number_of_open_complaints"].fillna("Unknown")
data ["vehicle_class"] = data ["vehicle_class"].fillna("Unknown")
data ["vehicle_size"] = data ["vehicle_size"].fillna("Unknown")
data ["vehicle_type"] = data ["vehicle_type"].fillna("Unknown")
data.isna().sum()
```
